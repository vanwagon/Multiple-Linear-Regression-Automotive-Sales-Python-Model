```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
#Loading data from the CSV file
data = pd.read_csv('advertising_data.csv')

```

```python
#setting variables
data['Month'] = pd.to_datetime(data['Month_year']).dt.month
data['Year'] = pd.to_datetime(data['Month_year']).dt.year
data['Budget'] = pd.to_numeric(data['Budget'])
data['Traffic_Visits'] = pd.to_numeric(data['Traffic_Visits'])
data['Shopping_Visits'] = pd.to_numeric(data['Shopping_Visits'])
data['Shopping_Rate'] = pd.to_numeric(data['shopping_Rate'])
```

```python
# Cost per traffic visit
cost_per_traffic_visit = data['Budget'].sum() / data['Traffic_Visits'].sum()
print('Cost per Traffic Visit:', cost_per_traffic_visit)

# Cost per shopping visit
cost_per_shopping_visit = data['Budget'].sum() / data['Shopping_Visits'].sum()
print('Cost per Shopping Visit:', cost_per_shopping_visit)
```

```python
#having two sets of data : one training, one testing
train, test = train_test_split(data, test_size=0.33, random_state=42)
```

```python
# training the model -- linear regression
traffic_model = LinearRegression()
traffic_model.fit(train[['Budget']], train['Traffic_Visits'])
```

```python
# trained to get shopping visits
shopping_model = LinearRegression()
shopping_model.fit(train[['Budget']], train['Shopping_Visits'])
```

```python
# trained to get shopping rate
rate_model = LinearRegression()
rate_model.fit(train[['Budget']], train['Shopping_Rate'])
```

```python
#make predictions
test['Traffic_Predicted'] = traffic_model.predict(test[['Budget']])
test['Shopping_Predicted'] = shopping_model.predict(test[['Budget']])
test['Rate_Predicted'] = rate_model.predict(test[['Budget']])
```

```python
#R-squared and mean squared error for each model
traffic_r2 = r2_score(test['Traffic_Visits'], test['Traffic_Predicted'])
shopping_r2 = r2_score(test['Shopping_Visits'], test['Shopping_Predicted'])
rate_r2 = r2_score(test['Shopping_Rate'], test['Rate_Predicted'])
```

```python
#showing mean average error, this is to see if how much error our model can generate on average

traffic_mae = mean_absolute_error(test['Traffic_Visits'], test['Traffic_Predicted'])
shopping_mae = mean_absolute_error(test['Shopping_Visits'], test['Shopping_Predicted'])
rate_mae = mean_absolute_error(test['Shopping_Rate'], test['Rate_Predicted'])
```

```python
# showing table with all metrics
table = pd.DataFrame({
    'Model': ['Traffic', 'Shopping', 'Rate'],
    'Formula': [f'y = {traffic_model.coef_[0]:.2f}x + {traffic_model.intercept_:.2f}', f'y = {shopping_model.coef_[0]:.2f}x + {shopping_model.intercept_:.2f}', f'y = {rate_model.coef_[0]:.4f}x + {rate_model.intercept_:.4f}'],
    'R-squared': [traffic_r2, shopping_r2, rate_r2],
    'Mean Absolute Error': [traffic_mae, shopping_mae, rate_mae]
})

print('Actual vs Predicted:')
display(test[['Month', 'Year', 'Traffic_Visits', 'Traffic_Predicted', 'Shopping_Visits', 'Shopping_Predicted', 'Shopping_Rate', 'Rate_Predicted']])
print('\n')
print('Model Metrics:')
display(table)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# plotting actual vs predicted for traffic visits
sns.regplot(x=test['Traffic_Predicted'], y=test['Traffic_Visits'], scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title('Actual vs Predicted Traffic Visits')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.show()

# plotting actual vs predicted for shopping visits
sns.regplot(x=test['Shopping_Predicted'], y=test['Shopping_Visits'], scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title('Actual vs Predicted Shopping Visits')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# plottingh actual vs predicted for shopping rate
sns.regplot(x=test['Rate_Predicted'], y=test['Shopping_Rate'], scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title('Actual vs Predicted Shopping Rate')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
